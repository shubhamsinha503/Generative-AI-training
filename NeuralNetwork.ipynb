# **Neural network of layer 1**




***keras*** is fucntion which learns pattern. Then we define the shape of the neural network with layers definition in sequence which is done by ***sequential***.


**Dense**   to define the layer of connected neuron.***Input Shape*** is used to define the shape of the input.

***Optimizer*** gueses the next guess for establishing the relationship bewteen x and y.

***Loss*** is used to define dataset of how was the guess how close was the guess




```python
model=tensorflow.keras.Sequential([tensorflow.keras.layers.Dense(units=1, input_shape=[1])])
model.compile(optimizer='sgd',loss='mean_squared_error')
```

```python
import tensorflow
import matplotlib
import seaborn
import numpy as np
```

***epochs*** is the loop which takes place where u let the whole process of optimization and loss fucntion go through all in specific batches.

```python
xs= np.array([-1,2,3,4,5,6],dtype=float)
ys= np.array([-21,122,33,42,35,16],dtype=float)
```

```python
model.fit(xs,ys,epochs=500)
```

***Here in the prediction , we give the value of x and get the result in the form of y.***

```python
 print(model.predict([10.0]))
```
